```python
import openpyxl

dosya = openpyxl.load_workbook("./Notlar.xlsx")
sayfa = dosya["BIL2001"]


sayfa.cell(row =1,column =1,value ="BAGIL DEGERLENDIRMEYE KATILMAYANLAR")
sayfa.cell(row =2,column =1,value ="SINIF ORTALAMASI")
sayfa.cell(row =3,column =1,value ="STANDART SAPMA")
sayfa.cell(row =4,column =1,value ="SINIF DUZEYI")
sayfa.cell(row =6,column =1,value ="NUMARA")
sayfa.cell(row =6,column =2,value ="VIZE")
sayfa.cell(row =6,column =3,value ="FINAL_NOTU")
sayfa.cell(row =6,column =4,value ="ORTALAMA")
sayfa.cell(row =6,column =5,value ="T NOTU")
sayfa.cell(row =6,column =6,value ="HARFLI NOT")


# Ortalama Sütunu
for satir in range(7,140):
    
    if sayfa.cell(satir,3).value != 0 and sayfa.cell(satir,2).value+sayfa.cell(satir,3).value >30 :
        vize = sayfa.cell(satir,2).value
        final  = sayfa.cell(satir,3).value
        ortalama = (vize + final) /2
        sayfa.cell(row = satir, column = 4 , value = ortalama )
       
        
    else:
        vize = sayfa.cell(satir,2).value
        final  = sayfa.cell(satir,3).value
        ortalama = (vize + final)*0
        sayfa.cell(row = satir, column = 4 , value = ortalama )

                        
# Bağıl Değerlendirmeye katılmayanlar 
count = 0
for row in sayfa.iter_rows(min_row=7, max_row=139, min_col=4, max_col=4):
    for cell in row:
        value = cell.value
        if value is not None and value == 0:
            count += 1
        
sayfa.cell(row = 1, column = 5, value = count)      
        

    
# Sınıf Ortalaması


values = []


for satir in range(7,140):
    
    if sayfa.cell(satir,4).value is not None and sayfa.cell(satir,4).value != 0:
        values.append(sayfa.cell(satir,4).value)
total = sum(values)
count = len(values)
average = total / count



sayfa.cell(row = 2, column = 5, value = average)      


# Standart sapma 

import statistics
        
veriler = []


for row in sayfa.iter_rows(min_row=7, max_row=139, min_col=4, max_col=4):
    for cell in row:
        if cell.value is not None and cell.value != 0:
            veriler.append(cell.value)


standart_sapma = statistics.stdev(veriler)


sayfa['E3'].value = standart_sapma





# T Notu

    
for satir in range(7,140):
    Sinif_Ortalamasi = sayfa.cell(row = 2 , column =5).value
    Standart_Sapma = sayfa.cell(row = 3 , column =5).value
    HBN = sayfa.cell(satir,4).value

    
    
    if sayfa.cell(satir,4).value != 0 and sayfa.cell(satir,3).value >= 45 :
       
        
        T_Notu = ((HBN - Sinif_Ortalamasi)/Standart_Sapma) *10+50
        sayfa.cell(row = satir, column = 5 , value = T_Notu )
    else:
            
        T_Notu = 0
        sayfa.cell(row = satir, column = 5 , value = T_Notu )
        

# Harfli Not
e2_value = sayfa['E2'].value
if isinstance(e2_value, (int, float)) and 47.5 <= e2_value <= 52.5:
    
    for row in sayfa.iter_rows(min_row=7, max_row=139, min_col=5, max_col=5):
        for cell in row:
            if cell.value < 32:
               
                sayfa.cell(row=cell.row, column=6).value = "FF"

e2_value = sayfa['E2'].value
if isinstance(e2_value, (int, float)) and 47.5 <= e2_value <= 52.5:
    
    for row in sayfa.iter_rows(min_row=7, max_row=139, min_col=5, max_col=5):
        for cell in row:
            if 32 <= cell.value < 36.99:
               
                sayfa.cell(row=cell.row, column=6).value = "FD"

                
                
e2_value = sayfa['E2'].value
if isinstance(e2_value, (int, float)) and 47.5 <= e2_value <= 52.5:
    
    for row in sayfa.iter_rows(min_row=7, max_row=139, min_col=5, max_col=5):
        for cell in row:
            if 37 <= cell.value < 41.99:
               
                sayfa.cell(row=cell.row, column=6).value = "DD"                
                
                
                
e2_value = sayfa['E2'].value
if isinstance(e2_value, (int, float)) and 47.5 <= e2_value <= 52.5:
   
    for row in sayfa.iter_rows(min_row=7, max_row=139, min_col=5, max_col=5):
        for cell in row:
            if 42 <= cell.value < 46.99:
                
                sayfa.cell(row=cell.row, column=6).value = "DC"                 
                
                
                
e2_value = sayfa['E2'].value
if isinstance(e2_value, (int, float)) and 47.5 <= e2_value <= 52.5:
   
    for row in sayfa.iter_rows(min_row=7, max_row=139, min_col=5, max_col=5):
        for cell in row:
            if 47 <= cell.value < 51.99:
                
                sayfa.cell(row=cell.row, column=6).value = "CC"       
                
                
                
e2_value = sayfa['E2'].value
if isinstance(e2_value, (int, float)) and 47.5 <= e2_value <= 52.5:
   
    for row in sayfa.iter_rows(min_row=7, max_row=139, min_col=5, max_col=5):
        for cell in row:
            if 52 <= cell.value < 56.99:
                
                sayfa.cell(row=cell.row, column=6).value = "CB"                       
                
                
e2_value = sayfa['E2'].value
if isinstance(e2_value, (int, float)) and 47.5 <= e2_value <= 52.5:
   
    for row in sayfa.iter_rows(min_row=7, max_row=139, min_col=5, max_col=5):
        for cell in row:
            if 57 <= cell.value < 61.99:
                
                sayfa.cell(row=cell.row, column=6).value = "BB"                

                
e2_value = sayfa['E2'].value
if isinstance(e2_value, (int, float)) and 47.5 <= e2_value <= 52.5:
    
    for row in sayfa.iter_rows(min_row=7, max_row=139, min_col=5, max_col=5):
        for cell in row:
            if 62 <= cell.value < 66.99:
               
                sayfa.cell(row=cell.row, column=6).value = "BA"      
                
                
e2_value = sayfa['E2'].value
if isinstance(e2_value, (int, float)) and 47.5 <= e2_value <= 52.5:
   
    for row in sayfa.iter_rows(min_row=7, max_row=139, min_col=5, max_col=5):
        for cell in row:
            if cell.value >= 67:
                
                sayfa.cell(row=cell.row, column=6).value = "AA"                      

                
#Sınıf Düzeyi

e4_value = sayfa['E4'].value

if isinstance(e2_value, (int, float)) and e2_value <= 42.5:
     sayfa.cell(row=4, column=5).value = "Zayıf"

elif isinstance(e2_value, (int, float)) and e2_value <= 42.5:
    sayfa.cell(row=4, column=5).value = "Kötü"

elif isinstance(e2_value, (int, float)) and 47.5 < e2_value <= 52.5:
    sayfa.cell(row=4, column=5).value = "Orta"

elif isinstance(e2_value, (int, float)) and 52.5 < e2_value <= 57.5:
    sayfa.cell(row=4, column=5).value = "Ortanın Üstü"
    
    
elif isinstance(e2_value, (int, float)) and 57.5 < e2_value <= 62.5:
    sayfa.cell(row=4, column=5).value = "İyi"    

    
elif isinstance(e2_value, (int, float)) and 62.5 < e2_value <= 70:
    sayfa.cell(row=4, column=5).value = "Çok İyi"    
    
elif isinstance(e2_value, (int, float)) and 70 < e2_value <= 80:
    sayfa.cell(row=4, column=5).value = "Mükemmel"    
    

else:
    sayfa.cell(row=4, column=5).value = "Üstün Başarı" 
    
    
    
                
dosya.save("./Notlar.xlsx")
```
